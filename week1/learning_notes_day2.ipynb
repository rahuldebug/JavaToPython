```python
#Java:

#Strict access control with keywords:
#public: accessible from anywhere.
#protected: accessible within package or subclasses outside package.
#private: only accessible within defining class.

#Python

class Car:
    def __init__(self, make,year):
        self.__make=make  # private attribute
        self._year=2025 #protected attribute
    def get_make(self):
        return self.__make
    def get_year(self):
        return self._year
my_car=Car("tata",2025)
print(my_car.get_make())
print(my_car.get_year())
```

```python
class Car:
    # Set the default value in the signature
    def __init__(self, make, year=2025): 
        self.__make = make
        self._year = year  # This now respects what is passed OR the default

    def get_make(self):
        return self.__make
        
    def get_year(self):
        return self._year

# Case 1: Using the Default
car_default = Car("Tata") 
print(f"Default Year: {car_default.get_year()}") # Output: 2025

# Case 2: Overriding the Default
car_custom = Car("Tata", 2026)
print(f"Custom Year: {car_custom.get_year()}")   # Output: 2026
```
