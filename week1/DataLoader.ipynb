```python
import sys
#!{sys.executable} -m pip install pandas
import pandas as pd
#print(f"Pandas version: {pd.__version__}")
from typing import Iterator

class DataScienceloader:
    def __init__(self , filepath: str ,chunk_size:int =1000):
        self.filepath=filepath
        self.chunk_size=chunk_size

    def stream_data(self) -> Iterator[pd.DataFrame]:
        #.  -> just tells ide what to expect , 
        for chunk in pd.read_csv(self.filepath, chunksize=self.chunk_size):
            # You can perform intermediate processing here before yielding
            # e.g., cleaning nulls, normalizing data
            cleaned_chunk = chunk.dropna() 
            yield cleaned_chunk
ds_loader = DataScienceloader("../data/data.csv", chunk_size=5)

for i, chunk in enumerate(ds_loader.stream_data()):
    print(f"Processing Chunk {i+1}...")
    
    # Do your work here (e.g., calculations, sending to a model)
    print(chunk.head())
```
