```python
def log_streamer():
    print("printing very long log file ")
    for i in range(100000):
        yield (f"printing line number {i}")
log = log_streamer ()
print(next (log)) 
print(next(log))
```

Quick Summary to Memorize:
[ ] = List. I can change it.

( ) = Tuple / Function. I cannot change it / I am executing something.

{ } = Dictionary / Set. I need fast lookups or unique items.</K,>

Way A: The Generator Function (yield)
When Python sees yield, it treats the function differently. Instead of returning a value and exiting, it freezes the function's state (all local variables) and hands the value back to the caller.

This is what we briefly touched on with brackets. Switching [] to () turns a list into a generator.

```python
massive_generator = (i for i in range (1000000))
print(next (massive_generator))
```
